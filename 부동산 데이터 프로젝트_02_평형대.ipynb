```python
gu_pyung = pd.crosstab(index = df['구'], columns =  df['평형대'])
gu_pyung = gu_pyung.reset_index()
gu_pyung
```

```python
fig = px.bar(gu_pyung, x = '구', y = ['18평 이하',	'18평-25평',	'25평-30평',	'30평-40평',	'40평 초과'], barmode = 'group')
fig.update_layout(template='simple_white', title_text='<b>구 별 거래량 (2006-2021년)</b>')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.update_layout(yaxis = dict(tickformat = "digits"))
fig.show()
```

```python
gu_pyung_ratio = gu_pyung.copy()

gu_pyung_ratio['sum'] = gu_pyung_ratio[['18평 이하', '18평-25평', '25평-30평', '30평-40평', '40평 초과']].sum(axis = 1)
gu_pyung_ratio['18평 이하 비율'] = ((gu_pyung_ratio['18평 이하'] / gu_pyung_ratio['sum']) * 100).round(0)
gu_pyung_ratio['18평-25평 비율'] = ((gu_pyung_ratio['18평-25평'] / gu_pyung_ratio['sum']) * 100).round(0)
gu_pyung_ratio['25평-30평 비율'] = ((gu_pyung_ratio['25평-30평'] / gu_pyung_ratio['sum']) * 100).round(0)
gu_pyung_ratio['30평-40평 비율'] = ((gu_pyung_ratio['30평-40평'] / gu_pyung_ratio['sum']) * 100).round(0)
gu_pyung_ratio['40평 초과 비율'] = ((gu_pyung_ratio['40평 초과'] / gu_pyung_ratio['sum']) * 100).round(0)

gu_pyung_ratio.drop(columns = {'18평 이하','18평-25평','25평-30평','30평-40평','40평 초과','sum'}, inplace = True)
gu_pyung_ratio = gu_pyung_ratio.round(2)
gu_pyung_ratio
```

#### 2-2 연도/평형대별 거래량

```python
year_pyung = pd.DataFrame(df.groupby(['계약연도','평형대'])['거래금액_억원'].count().reset_index())
year_pyung.rename(columns = {'거래금액_억원' : 'count'}, inplace = True)
year_pyung
```

```python
year_pyung_ratio = pd.pivot_table(year_pyung, index = ['계약연도'], columns = '평형대', values = ['count'])
year_pyung_ratio['sum'] = year_pyung_ratio.sum(axis = 1)
year_pyung_ratio['18평이하_비율'] =  (year_pyung_ratio['count']['18평 이하'] / year_pyung_ratio['sum']) * 100
year_pyung_ratio['18평-25평_비율'] = (year_pyung_ratio['count']['18평-25평'] / year_pyung_ratio['sum']) * 100
year_pyung_ratio['25평-30평_비율'] = (year_pyung_ratio['count']['25평-30평'] / year_pyung_ratio['sum']) * 100
year_pyung_ratio['30평-40평_비율'] = (year_pyung_ratio['count']['30평-40평'] / year_pyung_ratio['sum']) * 100
year_pyung_ratio['40평초과_비율'] = (year_pyung_ratio['count']['40평 초과'] / year_pyung_ratio['sum']) * 100
year_pyung_ratio = year_pyung_ratio.round(0)
year_pyung_ratio = year_pyung_ratio[['18평이하_비율','18평-25평_비율','25평-30평_비율','30평-40평_비율','40평초과_비율']]
year_pyung_ratio
```
