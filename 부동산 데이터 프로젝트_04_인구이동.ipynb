```python
# 전입신고 데이터에서 아파트 거래량을 뺀 데이터
아파트거래 = pd.DataFrame(df.groupby(['구'])['거래금액_억원'].agg('count').reset_index())
아파트거래.rename(columns = {'거래금액_억원' : '아파트거래량'}, inplace = True)
전입신고 = pd.DataFrame(df_인구이동.query('구간 == "총이동"').groupby(['자치구'])['계'].agg('sum').reset_index())
전입신고.rename(columns = {'계' : '전입신고수'}, inplace = True)

# 전입신고['기간'] = 전입신고['기간'].astype(int)
아파트_전입_비교 = pd.merge(아파트거래, 전입신고, left_on = ('구'), right_on = ('자치구'))
아파트_전입_비교['전입신고수_아파트제외'] = (아파트_전입_비교['전입신고수'] - 아파트_전입_비교['아파트거래량'])
아파트_전입_비교['아파트거래비율'] = (((아파트_전입_비교['아파트거래량'] / 아파트_전입_비교['전입신고수'])) * 100).round(2)
아파트_전입_비교['아파트제외거래비율'] = (((아파트_전입_비교['전입신고수_아파트제외'] / 아파트_전입_비교['전입신고수'])) * 100).round(2)
아파트_전입_비교
```

* 위의 전입신고에 따른 구별 총이동 차트와 아파트 실거래에 따른 구별 아파트 거래값이 차이가 나는 이유:
</br> '총이동' 차트의 데이터프레임은 아파트를 포함한 전세, 월세, 오피스텔, 빌라 등의 거래에서의 모든 전입신고 건수가 포함이 되었으나 
</br>아래의 차트는 아파트 거래의 값만 포함되어 있음

* 아래의 차트는 노원구->송파구->강남구->강동구->강서구 순으로 거래량이 많았으며
전입신고의 차트는 송파구->강남구->관악구->강서구->강동구 순으로 거래량이 많았다.
* 즉 '아파트를 포함한 '전세','월세','빌리' 등의 모든 주택 거래'는 [송파,강남,관악]에 집중되어 있으며 / '아파트 거래'는 [노원,송파,강남]에 집중되어 있음을 확인 

```python
아파트거래 = pd.DataFrame(df.groupby(['계약연도'])['거래금액_억원'].agg('count').reset_index())
아파트거래.rename(columns = {'거래금액_억원' : '아파트거래량'}, inplace = True)
전입신고 = pd.DataFrame(df_인구이동.query('구간 == "총이동"').groupby(['기간'])['계'].agg('sum').reset_index())
전입신고.rename(columns = {'계' : '전입신고수'}, inplace = True)

전입신고['기간'] = 전입신고['기간'].astype(int)
아파트_전입_비교 = pd.merge(아파트거래, 전입신고, left_on = ('계약연도'), right_on = ('기간'))
아파트_전입_비교['전입신고수_아파트제외'] = (아파트_전입_비교['전입신고수'] - 아파트_전입_비교['아파트거래량'])
아파트_전입_비교.drop(columns = {'기간'}, inplace = True)
아파트_전입_비교['아파트거래비율'] = (((아파트_전입_비교['아파트거래량'] / 아파트_전입_비교['전입신고수'])) * 100).round(2)
아파트_전입_비교['아파트제외거래비율'] = (((아파트_전입_비교['전입신고수_아파트제외'] / 아파트_전입_비교['전입신고수'])) * 100).round(2)
아파트_전입_비교
```

```python
df_시도간이동합 = pd.DataFrame(df_시도간이동.groupby('자치구')[df_인구이동.iloc[:,3:].columns].sum().reset_index())
df_시도간이동합
```

```python
data이동비율 = df_시도간이동합.copy()
data이동비율['직업']     = data이동비율['직업']     / data이동비율['계'] * 100
data이동비율['가족']     = data이동비율['가족']     / data이동비율['계'] * 100
data이동비율['주택']     = data이동비율['주택']     / data이동비율['계'] * 100
data이동비율['교육']     = data이동비율['교육']     / data이동비율['계'] * 100
data이동비율['주거환경'] = data이동비율['주거환경'] / data이동비율['계'] * 100
data이동비율['자연환경'] = data이동비율['자연환경'] / data이동비율['계'] * 100
data이동비율['기타']     = data이동비율['기타']     / data이동비율['계'] * 100
data이동비율 = data이동비율.round(1)
data이동비율 = data이동비율[['자치구','계', '주택','가족','직업','교육','기타','주거환경','자연환경']]
data이동비율
```
