#### 3-1 전체기간 상승비 (2006-2021년)

```python
# 전체기간 평형대 별 변동비

year_pyung = pd.DataFrame(df.groupby(['계약연도','평형대'])['평당금액_억원'].median().reset_index())
year_pyung['변동률'] = (year_pyung.sort_values('계약연도').groupby('평형대')['평당금액_억원'].pct_change() * 100).round(2)
year_pyung.loc[0:4, '변동률'] = 0
year_pyung['누적변화율'] = year_pyung.sort_values('계약연도').groupby('평형대')['변동률'].cumsum().round(2)
year_pyung[-5:][['계약연도','평형대','평당금액_억원','누적변화율']]
```

```python
fig = px.line(year_pyung, x = '계약연도', y = '누적변화율', color = '평형대', markers = True)
fig.update_layout(template='simple_white',  xaxis = dict(dtick = 1), title_text='<b>전체 기간 평당금액 누적 변화율 (2006-2021년)</b>')
fig.show()
```

```python
# 전체기간 상승비 (평당금액)

year_gu_price = pd.pivot_table(df, index = ['계약연도'], values = '평당금액_억원', columns = ['구'], aggfunc = 'mean')
year_gu_price = year_gu_price.round(2)
year_gu_price
```

```python
# 계약연도에 따라 표시된 거래금액에서 가장 첫번째 값을 기준으로 몇배가 올랐는 지 확인

year_gu_price_diff = year_gu_price.pct_change() + 1
year_gu_price_diff = year_gu_price_diff.cumprod()

year_gu_price_diff.loc[2006] = 1
year_gu_price_diff = year_gu_price_diff.round(2)
year_gu_price_diff
```

```python
data = year_gu_price_diff
fig = px.line(data, x = data.index, y = data.columns, markers = True)
fig.update_layout(template='simple_white',  xaxis = dict(dtick = 1), title_text='<b>전체 기간 평당금액 상승비 (2006-2021년)</b>')
fig.update_traces(textposition='top center')

fig.show()
```

##### 3-2-2 약세기 하락비

* 부동산의 강세기와 약세기를 구분할 수 있는 명확한 기준이 없기 때문에 </br>  </br>
*   2017-2021년 - 아파트 가격이 **뚜렷하게 상승한** 시기를 -> ***강세기***
*   2009-2013년 - 아파트 가격이 크게 상승하지 못하고 **하락/유지한** 시기를 -> ***약세기***</br> 로 구분하여 분석을 진행

```python
# 약세기 평형대 별 변동비

year_pyung = pd.DataFrame(df.query('계약연도 in [2009,2010,2011,2012,2013]').groupby(['계약연도','평형대'])['평당금액_억원'].median().reset_index())
year_pyung['변동률'] = (year_pyung.sort_values('계약연도').groupby('평형대')['평당금액_억원'].pct_change() * 100).round(2)
year_pyung.loc[0:4, '변동률'] = 0
year_pyung['누적변화율'] = year_pyung.sort_values('계약연도').groupby('평형대')['변동률'].cumsum().round(2)
year_pyung[-5:][['평형대','평당금액_억원','누적변화율']]
```

```python
fig = px.line(year_pyung, x = '계약연도', y = '누적변화율', color = '평형대', markers = True)
fig.update_layout(template='simple_white',  xaxis = dict(dtick = 1), title_text='<b>전체 기간 평당금액 누적 변화율 (2017-2021년)</b>')
fig.show()
```
