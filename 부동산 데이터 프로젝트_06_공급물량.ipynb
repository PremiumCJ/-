```python
누적세대수 = 공급물량.sort_values('입주년월', ascending = True)
누적세대수['누적세대수'] = 누적세대수.groupby(['구'])['총세대수'].apply(lambda x : x.cumsum())

# 입주가 진행되지 않은 입주연도가 있어 해당 입주연도의 값은 전년도와 동일한 값으로 채우기 위해 피봇테이블 형태로 변환하여 ffill을 사용
구별누적세대수 = pd.DataFrame(누적세대수.groupby(['구','입주연도'])['누적세대수'].max().reset_index())
구별누적세대수 = pd.pivot_table(구별누적세대수, index = '입주연도', columns = '구', values = '누적세대수')
구별누적세대수.loc[1986].fillna(0, inplace = True)
구별누적세대수 = 구별누적세대수.fillna(method = 'ffill', axis = 0)
구별누적세대수
```

```python
# ffill을 사용하여 모든 입주연도에 값을 채운 후 다시 데이터프레임 형태로 변환
누적세대수 = pd.melt(구별누적세대수.reset_index(), id_vars='입주연도', var_name='구', value_name='누적세대수')
누적세대수
```

```python
data = 누적세대수
data = pd.DataFrame(data.groupby('입주연도')['누적세대수'].sum().reset_index())
data.loc[15]['누적세대수'] - data.loc[0]['누적세대수']
data.loc[20]['누적세대수'] - data.loc[15]['누적세대수']
data['증가율'] = data['누적세대수'].pct_change() * 100
data.loc[0, '증가율'] = 0
data['증가율'] = data['증가율'].round(2)

세대수공급 = pd.DataFrame(공급물량.groupby('입주연도')['총세대수'].sum().reset_index())
세대수공급.drop(columns = {'입주연도'}, inplace = True)
세대수 = pd.concat([data, 세대수공급], axis = 1)
세대수
# px.bar(data, x = '입주연도', y = '누적세대수', color = '구')
```
